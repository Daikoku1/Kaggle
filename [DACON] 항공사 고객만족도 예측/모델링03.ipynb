```python
train[['Departure Delay in Minutes', 'Arrival Delay in Minutes']].corr()
```

```python
def change(x, y):
    if x == 0 :
        if y == 0 : return 'x=y=0'
        else: return 'A delay'
    else:
        if y == 0 : return 'D delay'
        else: return 'AD delay'

train['Delay'] = train.apply(lambda x : change(x['Departure Delay in Minutes'],x['Arrival Delay in Minutes']), axis = 1)
```

```python
print(train['Delay'].value_counts())
sns.barplot(data = train, x = 'Delay', y = 'target')
```

```python
def change(y):
    if y == 0 : return "A not delayed"
    if y < 3 : return "3 delayed"
    return "6 delayed"
train['Delay'] = train.apply(lambda x : change(x['Arrival Delay in Minutes']), axis = 1)
print(train['Delay'].value_counts())
sns.barplot(data = train, x = 'Delay', y = 'target')
```

```python
def change(y):
    if y == 0 : return 0
    if y < 3 : return 1
    return 2
# train data
train['Delay'] = train.apply(lambda x : change(x['Arrival Delay in Minutes']), axis = 1)
del train['Departure Delay in Minutes']
del train['Arrival Delay in Minutes']

test['Delay'] = test.apply(lambda x : change(x['Arrival Delay in Minutes']), axis = 1)
del test['Departure Delay in Minutes']
del test['Arrival Delay in Minutes']
```

##  
- pycaret 

```python
from pycaret.classification import *
reg = setup(data = train,
            silent = True,
            target = 'target',
            session_id = 0,
            n_jobs = 7)
```

```python
best = compare_models(n_select=3, fold = 10) #  3 
```

```python
blend_3_soft = blend_models(estimator_list=best, method='soft', choose_better = True)
predict_model(blend_3_soft).head(10)
```

```python
blend_3_soft = finalize_model(blend_3_soft)
prediction = predict_model(blend_3_soft, data = test)
```
